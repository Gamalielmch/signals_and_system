```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
a=0.1 
b=1
```

```python
x=np.linspace(0.01,1,100)
N = lambda x: 1000*np.exp(x)+ 435/x*(np.exp(x)-1)
fig=plt.figure(figsize=(3,3))
plt.plot(x,N(x))
plt.grid()
```

```python
Nx = lambda x: 1000*np.exp(x)+ 435/x*(np.exp(x)-1)-2000

for i in range(10):
    alpha=(b-a)/2+a
    Nx(alpha)
    if Nx(a)*Nx(alpha)>0:
        a=alpha
    else:
        b=alpha
print(alpha, Nx(alpha))
```

```python
fig=plt.figure(figsize=(3,3))
plt.plot(x,N(x))
plt.axhline(y=N(alpha),color='k')
plt.grid()
# plt.ylim(1500,3000)
```
