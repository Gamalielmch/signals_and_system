```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def coef(k,alpha):
    if np.abs(k)!=1:
        parte1=(1/(2*np.pi*(k**2-1) ))
        parte2= np.exp(-1j*k*alpha)*(np.cos(alpha)+1j*k*np.sin(alpha))
        parte3= (-1)**(k+1)
        final=parte1*(parte3-parte2)
    elif k==-1:
        final=(-1/(8*np.pi))* (1-np.cos(2*alpha)-1j*(2*np.pi-2*alpha+np.sin(2*alpha)))
    else:
        final=(-1/(8*np.pi))* (1-np.cos(2*alpha)+1j*(2*np.pi-2*alpha+np.sin(2*alpha)))
    return final
```

```python
coeficientes=[]
nc=50
k=np.arange(-nc,nc+1)
for i in k:
    coeficientes.append(coef(i.astype(float),np.pi/2))
plt.stem(np.abs(coeficientes))
```
