```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-np.pi, np.pi,50)
fx= 40+x**3-np.sin(3*x)-6*np.cos(2*x)
plt.plot(x,fx)
```

```python
ni=30
xn=np.linspace(-np.pi, np.pi,ni)
val=40+xn**3-np.sin(3*xn)-6*np.cos(2*xn)
wt=((2*np.pi)/ni )/ (ni-1)
w=(2*np.pi)/ni +wt
plt.bar(xn,val,width=w,edgecolor='r')
plt.plot(x,fx,'k')
plt.show()
```
