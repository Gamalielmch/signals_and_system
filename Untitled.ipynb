```python
import numpy as np 
import librosa 
import matplotlib.pyplot as plt 
import IPython.display as ipd
from scipy.fft import fft, fftfreq
```

```python
x,sr1=librosa.load('vocal_a.mp3')
ipd.Audio(x,rate=sr1)
plt.plot(x[15000:16000])
```

```python
Fs = sr1
T = 1 / Fs
N =len(x)
#### Eje de frecuencias
xf = fftfreq(N, T)
xf= np.fft.fftshift(xf)
##### Transformada rápida de Fourier
x_fourier = np.fft.fft(x[:])
x_shift = np.fft.fftshift(x_fourier)
fig=plt.figure(figsize=(15,4))
plt.plot(xf,np.abs(x_shift))
```
