```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
figure, ax = plt.subplots(figsize=(3, 3))
ax.plot(2,2,'sb')
ax.set_xlim(-3,3)
ax.set_ylim(-3,3)
ax.grid()
plt.axhline(y = 0, color = 'k', linestyle = '-')
plt.axvline(x = 0, color = 'k', linestyle = '-')
plt.xlabel('Real')
plt.ylabel('Imaginario')
real2=np.sqrt(8)*np.cos(np.deg2rad(45))
img2=np.sqrt(8)*np.sin(np.deg2rad(45))
ax.plot(real2,img2,'*r')
plt.show()
```
