```python
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(6, 11))
ax1 = plt.subplot(321)
x=np.array([0,0,0,1,2,2,2,2,0])
n=np.array([-2,-1,0,1,2,3,4,5,6])
plt.stem(n,x)
ax1.title.set_text('Señal original')

ax2 =plt.subplot(322)
plt.stem(-n,x)
ax2.title.set_text('Señal invertida')


ax3 =plt.subplot(323)
plt.stem(n+2,x)
ax3.title.set_text('Señal retrasada')


ax4 =plt.subplot(324)
plt.stem(n-1,x)
ax4.title.set_text('Señal adelantada')

ax5 =plt.subplot(325)
plt.stem(2*n,x)
ax5.title.set_text('Señal expandida')


ax6 =plt.subplot(326)
plt.stem(n/2,x)
ax6.title.set_text('Señal comprimida')
```

```python
t=np.linspace(-4*np.pi,4*np.pi,200)
x=np.sin(t)
plt.plot(t,x, label='Función seno')
plt.grid()
plt.axvline(x=0,color='k')
plt.axhline(y=0,color='k')
plt.legend(loc="upper right")
plt.title('Grafica de seno');
```

```python
x = lambda t: np.sin(t)
def Tr(x):
    return lambda t: 2*x(t)

t = np.linspace(-5,5,300)
plt.plot(t, x(t), label='Input $x(t)$')
y = Tr(x)
plt.plot(t, y(t), label='Output $y(t)=2x(t)$')
plt.legend(loc="upper right")
plt.grid()
plt.title('Sistema');
```

```python
def Tr(x):
    return lambda t: x(t)**2 # Este es el sistema
t = np.linspace(-10, 10, 300)
x1 = lambda t: np.sin(t)
x2 = lambda t: (t % (2*np.pi))/np.pi
x1plusx2 = lambda t: x1(t)+x2(t)
plt.figure(figsize=(6,3))
plt.plot(t,Tr(x1)(t)+Tr(x2)(t),'b')
plt.plot(t,Tr(x1plusx2)(t),'r--')
```

```python
def Tr(x):
    return lambda t: t*x(t)
t = np.linspace(-2,4, 300)
x = lambda t: np.exp(-t)*(t>=0).astype(float)
y1 = Tr(x)
y2 = Tr(lambda t: x(t-0.5))
y3 = Tr(lambda t: x(t-1))
plt.plot(t,y1(t),'r',t,y2(t),'g',t,y3(t),'b')
```
